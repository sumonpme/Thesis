```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.lines as mlines

file_path = 'As_Sn_recovery.xlsx'
df = pd.read_excel(file_path)

titles = [
    'pH2_Phizs Extra 500g/t_Protelan LS9011 500g/t',
    'pH2_Phizs Extra 500g/t_Protelan LS9011 2000g/t',
    'pH2_Phizs Extra 2000g/t_Protelan LS9011 500g/t',
    'pH2_Phizs Extra 2000g/t_Protelan LS9011 2000g/t',
    'pH4_Phizs Extra 1250g/t_Protelan LS9011 1250g/t',
    'pH6_Phizs Extra 500g/t_Protelan LS9011 500g/t',
    'pH6_Phizs Extra 500g/t_Protelan LS9011 2000g/t',
    'pH6_Phizs Extra 2000g/t_Protelan LS9011 500g/t',
    'pH6_Phizs Extra 2000g/t_Protelan LS9011 2000g/t'
]

# Define color, linestyle, and marker for each block
block_styles = [
    ('darkorange', '-', 'o'),  # pH2, 500, 500
    ('darkorange', '-', 's'),  # pH2, 500, 2000
    ('darkorange', ':', 'o'),  # pH2, 2000, 500
    ('darkorange', ':', 's'),  # pH2, 2000, 2000
    ('red', '--', 'd'),        # pH4, 1250, 1250
    ('royalblue', '-', 'o'),   # pH6, 500, 500
    ('royalblue', '-', 's'),   # pH6, 500, 2000
    ('royalblue', ':', 'o'),   # pH6, 2000, 500
    ('royalblue', ':', 's')    # pH6, 2000, 2000
]

fig, axs = plt.subplots(3, 3, figsize=(12, 12))
fig.suptitle('Selectivity of As against Sn', fontsize=18, fontweight='bold')

for block in range(9):
    ax = axs[block // 3, block % 3]
    color, linestyle, marker = block_styles[block]

    for test in range(3):
        as_col = f'Grade_B{block+1}_T{test+1}'
        sn_col = f'Rec_B{block+1}_T{test+1}'

        ax.plot(df[as_col], df[sn_col],
                linestyle=linestyle,
                marker=marker,
                color=color,
                label=f'Test_{test+1}',
                linewidth=1,
                markersize=4)

    # Add 45-degree reference line
    ax.plot([0, 100], [0, 100], 'k-', linewidth=0.5)
    ax.set_title(titles[block], fontsize=10)
    ax.set_xlim(0, 80)
    ax.set_ylim(0, 80)

    if block % 3 == 0:
        ax.set_ylabel('Cumulative As Recovery %', fontsize=10)
    if block // 3 == 2:
        ax.set_xlabel('Cumulative Sn Recovery %', fontsize=10)

# Custom legend elements

# Color (pH)
legend_pH = [
    mlines.Line2D([], [], color='darkorange', lw=2, label='pH 2'),
    mlines.Line2D([], [], color='red', lw=2, label='pH 4'),
    mlines.Line2D([], [], color='royalblue', lw=2, label='pH 6')
]

# Line style (Phizs Extra dosage)
legend_phizs = [
    mlines.Line2D([], [], color='black', linestyle='-', lw=2, label='Phizs Extra 500g/t'),
    mlines.Line2D([], [], color='black', linestyle='--', lw=2, label='Phizs Extra 1250g/t'),
    mlines.Line2D([], [], color='black', linestyle=':', lw=2, label='Phizs Extra 2000g/t')
]

# Marker (Protelan LS9011 dosage)
legend_protelan = [
    mlines.Line2D([], [], color='black', marker='o', linestyle='None', label='Protelan LS9011 500g/t'),
    mlines.Line2D([], [], color='black', marker='d', linestyle='None', label='Protelan LS9011 1250g/t'),
    mlines.Line2D([], [], color='black', marker='s', linestyle='None', label='Protelan LS9011 2000g/t')
]

# Place legends below the plots
fig.legend(handles=legend_pH + legend_phizs + legend_protelan,
           loc='lower center', ncol=3, fontsize=10, frameon=False, bbox_to_anchor=(0.5, -0.02))

plt.tight_layout(rect=[0, 0.05, 1, 0.95])
plt.savefig('selectivity_plot_with_legend.png', dpi=300, bbox_inches='tight')
plt.show()
```
