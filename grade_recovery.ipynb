```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

# Load Excel file
file_path = 'Fe_grade_recovery.xlsx'
df = pd.read_excel(file_path)

# Titles for each subplot
titles = [
    'pH2_Phizs Extra 500g/t_Protelan LS9011 500g/t',
    'pH2_Phizs Extra 500g/t_Protelan LS9011 2000g/t',
    'pH2_Phizs Extra 2000g/t_Protelan LS9011 500g/t',
    'pH2_Phizs Extra 2000g/t_Protelan LS9011 2000g/t',
    'pH4_Phizs Extra 1250g/t_Protelan LS9011 1250g/t',
    'pH6_Phizs Extra 500g/t_Protelan LS9011 500g/t',
    'pH6_Phizs Extra 500g/t_Protelan LS9011 2000g/t',
    'pH6_Phizs Extra 2000g/t_Protelan LS9011 500g/t',
    'pH6_Phizs Extra 2000g/t_Protelan LS9011 2000g/t'
]

# Style mapping: (marker for Protelan, linestyle for Phizs Extra)
style_map = {
    0: ('o', '-'),  # pH2, 500g/t Protelan, 500g/t Phizs
    1: ('s', '-'),  # pH2, 2000g/t Protelan, 500g/t Phizs
    2: ('o', ':'),  # pH2, 500g/t Protelan, 2000g/t Phizs
    3: ('s', ':'),  # pH2, 2000g/t Protelan, 2000g/t Phizs
    4: ('^', '--'), # pH4, 1250g/t both
    5: ('o', '-'),  # pH6, 500g/t Protelan, 500g/t Phizs
    6: ('s', '-'),  # pH6, 2000g/t Protelan, 500g/t Phizs
    7: ('o', ':'),  # pH6, 500g/t Protelan, 2000g/t Phizs
    8: ('s', ':'),  # pH6, 2000g/t Protelan, 2000g/t Phizs
}

# pH color function
def get_color(index):
    if index < 4:
        return 'darkorange'  # pH 2
    elif index == 4:
        return 'red'         # pH 4
    else:
        return 'royalblue'   # pH 6

# Create subplots
fig, axs = plt.subplots(3, 3, figsize=(18, 12))
fig.suptitle('Cumulative Fe₂O₃ Grade-Recovery Curve', fontsize=18, fontweight='bold')

# Plot each block
for block in range(9):
    ax = axs[block // 3, block % 3]
    color = get_color(block)
    marker, linestyle = style_map[block]

    for test in range(3):
        grade_col = f'Grade_B{block+1}_T{test+1}'
        rec_col = f'Rec_B{block+1}_T{test+1}'

        ax.plot(df[grade_col], df[rec_col],
                linestyle=linestyle,
                marker=marker,
                color=color,
                linewidth=0.75,
                markersize=4)

    # Titles and limits
    ax.set_title(titles[block], fontsize=10)
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 80)
    ax.set_xticks([0, 2, 4, 6, 8, 10])
    ax.set_yticks([0, 20, 40, 60, 80])

    # Show labels only on leftmost and bottom plots
    if block % 3 == 0:
        ax.set_ylabel('Cum. Fe Recovery (%)', fontsize=10)
    else:
        ax.set_yticklabels([])

    if block // 3 == 2:
        ax.set_xlabel('Cum. Fe Grade (%)', fontsize=10)
    else:
        ax.set_xticklabels([])

# Custom legend
legend_elements = [
    Line2D([0], [0], color='darkorange', lw=2, label='pH 2'),
    Line2D([0], [0], color='red', lw=2, label='pH 4'),
    Line2D([0], [0], color='royalblue', lw=2, label='pH 6'),
    Line2D([0], [0], color='black', linestyle='-', label='Phizs Extra 500g/t'),
    Line2D([0], [0], color='black', linestyle='--', label='Phizs Extra 1250g/t'),
    Line2D([0], [0], color='black', linestyle=':', label='Phizs Extra 2000g/t'),
    Line2D([0], [0], color='black', marker='o', linestyle='None', label='Protelan LS9011 500g/t'),
    Line2D([0], [0], color='black', marker='^', linestyle='None', label='Protelan LS9011 1250g/t'),
    Line2D([0], [0], color='black', marker='s', linestyle='None', label='Protelan LS9011 2000g/t'),
]

fig.legend(handles=legend_elements, loc='center right', fontsize=10)

plt.tight_layout(rect=[0, 0.03, 0.9, 0.95])  # Adjust for legend
plt.show()
```
